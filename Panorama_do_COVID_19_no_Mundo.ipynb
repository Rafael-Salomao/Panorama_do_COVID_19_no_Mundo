# Panorama do COVID-19 no Mundo

No ano de 2020, a humanidade se deparou com um desafio descomunal. A pandemia causada pelo Covid-19 surpreendeu o mundo com sua vertiginosa capacidade de transmissão. O vírus se espalhou rapidamente por todos os cantos do mundo gerando impactos sanitários, econômicos e sociais em todo o globo.

A Covid-19 é uma doença causada pelo vírus SARS-COV-2, a sigla SARS diz respeito à Síndrome Respiratória Aguda Grave. A doença foi identificada no fim de 2019, na cidade chinesa de Wuhan, e logo avançou para países de todos os continentes.

Inicialmente, o vírus infectou morcegos e, posteriormente, foi transmitida aos seres humanos. A Covid-19 ataca os pulmões do indivíduo, gerando quadros severos de insuficiência respiratória que podem levar ao óbito. A doença começa com sintomas de uma gripe comum e pode avançar para quadros de pneumonias graves.

### Valores ausentes

Com os outputs encontrados, já é possivel perceber a necessidade de tratamentos na base de dados. No entanto, antes de realizar as alterações e análises mais aprofundadas, é necessário checar a existência de valores ausentes ou nulos no Dataset.

Quando se trata de explorar os dados de um dataframe, uma das principais etapas é avaliar os valores nulos presentes. Devido a diferentes métodos no momento da coletas dos dados, é muito comum que haja erro de preenchimento, ausência de informação, campos em branco e até mesmo valores irregulares ou desconhecidos em um dataset. À esses dados, é atribuido o nome de valores nulos. 

Tais valores não são desejados em análises de Ciência de Dados, pois podem afetar o desempenho de certas ferramentas: algoritmos de Machine Learning, por exemplo. Para este presente artigo, os valores nulos serão tratados e manipulados para atender ao nosso objetivo, que é: obter os gráficos desejados à respeito do panorama do Covid-19 no mundo.

Em virtude do que foi dito, primeiramente, é preciso avaliar a quantidade e distribuição de tais valores nulos, presentes no Dataset. Abaixo pode-se observar a quantidade e a porcentagem de valores nulos em cada uma das colunas.

```
# Mostrar a quantidade de valores nulos em cada coluna, em ordem decrescente.
print('A quantidade de valores nulos em cada coluna é a seguinte:')
print(df.isnull().sum().sort_values(ascending = False))

# Mostrar a parcela de valores nulos em cada coluna, em ordem decrescente.
print('\nA porcentagem de valores nulos em cada coluna é a seguinte:')
print((df.isnull().sum()/df.shape[0]).sort_values(ascending = False))
```

É possível observar que há valores nulos na maior parte das colunas, isso pode ser um problema - como dito anteriormente - no entanto, é preciso avalia-los e entender o motivo da presença deles.

Para inferir sobre os dados nulos, é preciso voltar a consciência situcional do dataset. 

A partir das visualização iniciais da base de dados, é possível observar que a coluna `date` representa os dias correspondentes à coleta das informações das outras variáveis.

Com este insight, é possivel deduzir, que nos dias iniciais ou até mesmo outros dias em que apresentam valores nulos, não foram coletadas informações - na localidade em questão - ou até mesmo não houveram dados a serem computados naquele dia. 

Para ilustrar melhor esta condição, serão utilizados como exemplo os registros referentes ao Afeganistão. Na coluna `total_deaths`, os dados registrados, começaram a ser computador no dia `24-2-2020`. No entanto, o primeiro registro de morte ocorreu apenas no dia `23-3-2020`. Com isso, as informações referentes a ocorrência de mortes na variável `total_deaths`, no periodo anterior ao registro da primeira ocorrência, constam como nulos.

Entretanto, apesar desses valores constarem como nulos, são importantes para a análise e apresentação dos gráficos. Ou seja, é relevante, pois, para os dias em que não apresentaram mortes - e constam valores nulos nesta variável -, contém informações relevantes em outras variáveis, como por exemplo, a ocorrência de novos casos durante o mesmo período.

Portanto, a partir da análise dos valores ausentes, é possível avaliar, que tais informações fazem sentido estarem nulos. Contudo, é necessário substituir esses elementos ausentes por elementos não nulos, ou seja, os conjuntos que constam sem informação serão substituído pelo numeral "zero". Assim, torna-se possível a estrututação dos gráficos e manipulação das informações de modo mais eficiente através do Python. 

### Tratamento dos dados

#### Valores ausentes

Por fim, após a avaliação dessas entradas indesejadas no dataset, é preciso realizar o tratametento das informações.
```
# Criar uma cópia do dataset original.
df_copy = df.copy()

# Selecionar as colunas do dataset a serem analisadas.
df_copy = df[['iso_code', 'continent', 'location','date',
             'total_cases', 'new_cases','new_cases_smoothed', 
             'total_deaths','new_deaths', 'new_deaths_smoothed', 
             'total_cases_per_million','new_cases_per_million','new_cases_smoothed_per_million',
             'total_deaths_per_million','new_deaths_per_million','new_deaths_smoothed_per_million',
             'people_vaccinated', 'people_fully_vaccinated', 'new_vaccinations', 'population',
              'gdp_per_capita', 'extreme_poverty','hospital_beds_per_thousand','life_expectancy']]

# Preencher os valores nulos da coluna 'people_vaccinated'.
df_copy.people_vaccinated.fillna( 0, axis = 0, inplace = True)

# Preencher os valores nulos da coluna 'people_fully_vaccinated'.
df_copy.people_fully_vaccinated.fillna( 0, axis = 0, inplace = True)

# Preencher os valores nulos da coluna 'new_vaccinations'.
df_copy.new_vaccinations.fillna( 0, axis = 0, inplace = True)
```

```
# Mostrar o tipo das variáveis (colunas).
print('Tipos de variáveis:\n{}'.format(df_copy.dtypes))

# Transformar a coluna 'date' no formato 'datetime'
df_copy.date = pd.to_datetime(df_copy.date, format='%Y-%m-%d')

# Mostrar os valores da coluna 'date' ordenados
print('Coluna `date` em ordem crescente:\n{}'.format(df_copy.date.dt.date.sort_values()))

# Mostrar a data mais antiga e mais recente presente no dataset
print('\nO dataset contempla dados desde o dia {} até {}.'.format(df_copy.date.dt.date.min(),df_copy.date.dt.date.max()))

# Remover as linhas nulas na coluna 'continent' - referentes aos dados que regiões diferentes de países.
df_copy_apenas_paises = df_copy.dropna(subset=['continent'])
```

```
#Transforma os tick de valores altos (Bilhões, milhões e milhares) em abreviações B, M e K.
def format_tick(tick_val, pos):
    if tick_val >= 1000000000:
        val = round(tick_val/1000000000, 1)
        new_tick_format = '{:}B'.format(val)
    elif tick_val >= 1000000:
        val = round(tick_val/1000000, 1)
        new_tick_format = '{:}M'.format(val)
    elif tick_val >= 1000:
        val = round(tick_val/1000, 1)
        new_tick_format = '{:}K'.format(val)
    elif tick_val < 1000:
        new_tick_format = round(tick_val, 1)
    else:
        new_tick_format = tick_val

    # Transformar new_tick_format em string.
    new_tick_format = str(new_tick_format)
    
    # Remove o "0" após o ponto decimal.
    index_of_decimal = new_tick_format.find(".")
    
    if index_of_decimal != -1:
        value_after_decimal = new_tick_format[index_of_decimal+1]
        if value_after_decimal == "0":
          
            # Remove o 0 após o ponto.
            new_tick_format = new_tick_format[0:index_of_decimal] + new_tick_format[index_of_decimal+2:]
            
    return new_tick_format
```

```
#plot gráficos
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=((12,5)))

top_5_cases.plot(x='location',y='total_cases',kind='bar',ax=ax1)
ax1.set_title("Top 5 maior número de casos\n")
ax1.set_ylabel("Total mortes")
ax1.set_xlabel("País")
ax1.yaxis.set_major_formatter(tick.FuncFormatter(format_tick))

top_5_deaths.plot(x='location',y='total_deaths',kind='bar',ax=ax2)
ax2.set_title("Top 5 maior número de mortes\n")
ax2.set_ylabel("\n\n\n\nTotal mortes")
ax2.set_xlabel("País")
ax2.yaxis.set_major_formatter(tick.FuncFormatter(format_tick))

plt.tight_layout()
```

### Análise socio-economicas dos impactos da pandemia

O dataset utilizado também inclui informações sobre as situações socio-economicas das regiões apresentadas. A fim de tirar proveito dessas bases e presumir hipoteses a serem discutidas, será feita uma apresentação comparativa entre elementos como PIB per capita e expectativa de vida dos países.

Para tal observação são usados gráficos de dispersão. Também conhecido como Scatter Plot (nomenclatura em inglês), usa pontos para representar valores para duas variáveis ​​numéricas diferentes. Sendo que, a posição de cada ponto no eixo horizontal e vertical indica valores para um ponto determinado individual. 

Os gráficos de dispersão são usados ​​para observar as relações entre as variáveis, uma vez que, os pontos em um gráfico de dispersão, não apenas relatam os valores de pontos de dados individuais, mas também padrões quando os dados são tomados como um todo.

Mais informações sobre o conceito e estrutura dos gráficos de dispersão pode-se encontrar [neste link](https://chartio.com/learn/charts/what-is-a-scatter-plot/).

A apresentação dos gráficos de dispersão será feita pela biblioteca `Plotly Express`. Esta se refere a uma interface do `Plotly` de alto nível e usuais, que opera em uma variedade de tipos de dados e possibilita refinamento descomplicado das figuras ilustradas.

Os gráficos são elaborados de modo que o eixo x se trata dos valores sócio-economicos e o eixo y refere-se aos dados pandêmicos registrados por milhão de habitantes. Sendo que os pontos marcados diz respeito aos países indicados enquanto suas dimensões são equivalentes a suas populações totais.

## Resultados e discussões

### Interpretação correlação e causalidade

Considerando a afirmação acima juntamente aos resultados observados, é importante se ter cuidado ao analisar gráficos referentes a correlação de variáveis. Correlação não implica causalidade.

O problema não é o gráfico em si, mas sua interpretação. A relação entre a dispersão dos valores encontrados a partir de duas variáveis, não implica que uma variável é - necessariamente - responsável pela mudança na outra variável. É possível que exista uma terceira variável influenciando a relação ou que seja uma mera coincidência.

Tendo os valores de dispersão elencados nos gráficos acima, é possível inferir que países com melhores condições socio-econômicos foram mais impactados pela pandemia - apresentando maiores índices de mortalidade e contaminação. 

Tal conclusão chega a ser contraditória, ao ponto que, países com melhores condições econômicas tendem a ter melhores condições de saúde pública, acesso a informação, políticas públicas para evitar contaminações, dentre outros fatores que contribuem com a disseminação de doenças. No entanto, os gráficos mostram um maior impacto nos países mais ricos.

Esta interpretação é um exemplo de que correlação não implica causalidade. Os maiores índices de PIB per capita e expectativa de vida não são as causas dos elevados registros de casos e mortes ao redor do mundo.

### Hipóteses

Com o que foi dito, pode-se elencar hipóteses sobre motivos que levaram países de maior poder econômico serem mais efetados pela pandemia. 
- Países com pessoas de maior poder aquisitivo tendem a viajar mais e consequentemente pode aumentar a velocidade de contágio da doença. Por outro lado, regiões menos desenvolvidas financeiramente tendem a ser países menos globalizados e com contato abreviado entre diferentes comunidades.
- Regiões com cenários econômicas mais consolidados são mais propensas a terem grandes shows e concertos de entretenimento e música, que atraem vultosos públicos e representam um fator de aumento ao contágio da pandemia. 
- Países com melhores condições sociais possuem sistemas para registro de informações mais sofisticados e confiáveis, o que permite o registro mais fidedigno dos números reais ocorridos.

Tais hipóteses podem ser exploradas e analisadas em futuros projetos a fim de aferir a reais relação entre as variáveis e os números pandêmicos registrados.
